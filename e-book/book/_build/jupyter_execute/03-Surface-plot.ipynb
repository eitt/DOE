# The Usefulness of Surface Plots in Data Analysis for the Design of Experiments
```{contents}
```
Data visualization plays a pivotal role in data analysis, especially when it comes to the design of experiments. One form of visualization stands out in this domain: the surface plot. Surface plots provide a valuable tool for data analysis within the design of experiments. By allowing researchers to visualize the interactions between variables, guide the design process, facilitate response optimization, and enhance the understanding of complex phenomena, surface plots can significantly contribute to effective and insightful experimentation. This three-dimensional plotting technique visually represents a function of two variables: the x and y-axes are the inputs, and the z-axis is the output. This article explores the benefits and insights that surface plots can offer in experimental design and data analysis.



## Visualizing Interactions Between Variables

One of the principal benefits of surface plots is their ability to illustrate the interactions between two independent variables and their combined effect on a dependent variable. In the context of experimental design, this can help researchers understand how different factors (independent variables) interact to affect the outcome (dependent variable). The interaction usually generates a change in the slope of the surface, increasing or decreasing the function value. An important concept is curvature which refers to the nonlinear relationship between the response variable and one or more factors. Curvature can depend on several factors, including interactions between factors, nonlinear effects of individual factors, and measurement error.

For instance, imagine a design of an experiment with two factors, Temperature, and pressure affecting the yield of a chemical reaction. A surface plot can show how the combination of Temperature and pressure impacts the yield. By observing the plot, researchers can identify optimal operating conditions and visually comprehend the interactions between these factors.



## Guiding the Design of Experiments

Surface plots can play an instrumental role in the experimental design process. By offering a visual summary of the relationship between variables, they can guide researchers in selecting appropriate levels for experimental factors, leading to more efficient experimentation and more reliable results.

For example, researchers might use a surface plot to identify a region of factor settings (i.e., Temperature and pressure) that maximizes the yield. In subsequent experiments, they could focus on this region to fine-tune factor settings and achieve the highest possible yield.



## Facilitating Response Optimization

In many experimental design scenarios, the goal is to optimize a response â€” to find the factor levels that minimize or maximize a certain outcome. Surface plots can aid tremendously in this process. The 'peaks' and 'valleys' on the surface plot illustrate where the response reaches its maximum and minimum values, respectively. This visualization can facilitate the identification of optimal factor settings and thus streamline the optimization process.



## Enhancing Understanding of Complex Phenomena

Finally, surface plots can enhance our understanding of complex phenomena. The three-dimensional nature of these plots provides a rich, detailed view of the relationships between variables. This visual representation helps elucidate complex interactions that may be difficult to understand through numbers or tables alone.

# Visual examples

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

## Example 1. Yield is independent of Temperature and pressure.
In this case, the yield will be constant no matter what the temperature and pressure values are. We'll arbitrarily set the yield to be 10: $yield = 10$.

````python
# Create data
temperature = np.linspace(0, 100, 100)
pressure = np.linspace(0, 10, 100)
temperature, pressure = np.meshgrid(temperature, pressure)

yield_value = np.full_like(temperature, 10)

# Create surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(temperature, pressure, yield_value)
ax.set_xlabel('Temperature')
ax.set_ylabel('Pressure')
ax.set_zlabel('Yield')
plt.show()
````

## Example 2. Yield depends on Temperature and pressure but is linear.
In this case, let's assume $yield = 2 * temperature + 3 * pressure$.

````python
# Create data
yield_value = 2 * temperature + 3 * pressure

# Create surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(temperature, pressure, yield_value)
ax.set_xlabel('Temperature')
ax.set_ylabel('Pressure')
ax.set_zlabel('Yield')
plt.show()
````

## Example 3. Yield depends on Temperature and pressure, but there is a curvature (pressure*temperature).
In this case, $yield = temperature * pressure.$

````python
# Create data
yield_value = temperature * pressure

# Create surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(temperature, pressure, yield_value)
ax.set_xlabel('Temperature')
ax.set_ylabel('Pressure')
ax.set_zlabel('Yield')
plt.show()
````

## Example 4. Yield depends on Temperature and pressure, but there is a nonlinear nonconvex relationship.
In this case, we can create a non-linear, non-convex relationship using a sin function: $yield = sin(temperature) * pressure$.

````python
# Create data
yield_value = np.sin(temperature) * pressure

# Create surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(temperature, pressure, yield_value)
ax.set_xlabel('Temperature')
ax.set_ylabel('Pressure')
ax.set_zlabel('Yield')
plt.show()
````

## Final remarks

In the real world, the relationships between factors and response variables are often far more complex than the simplified examples I gave. For example, the yield of a chemical reaction may depend on not only the Temperature and pressure but also the concentration of the reactants, the presence of impurities, and the reaction time. In addition, real-world relationships are often stochastic, meaning they are not perfectly deterministic. Meaning that even if we know all of the factors that affect a response variable, we cannot perfectly predict the value of the response variable. Despite the complexity and stochasticity of real-world relationships, DOE can still be a valuable tool for understanding and optimizing these relationships. By carefully designing experiments and using statistical analysis, DOE practitioners can identify the factors that most affect a response variable and the optimal settings for these factors.
